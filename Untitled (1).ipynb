#### 4. Проверка однозначности ветвления по первому символу
```
1.  <V> ::= <F>(<V>)<D> | <P><D>
    v1 = <F>(<V>)<D>    v2=<P><D>
    L(v1) = {f, g}      L(v2) = {a, b, c, d}
    L(v1) ∩ L(v2) = ∅
    
2.  <D> ::= <Z><V> | Λ 
    d1 = <Z><V>             d2 = Λ
    L(d1) = {+, -, *, /}    L(d2) = {Λ}
    L(d1) ∩ L(d2) = ∅
    
3.  <Z> ::= + | - | * | / 
    z1 = +        z2 = -        z3 = *        z4 = /
    L(z1) = {+}   L(z2) = {-}   L(z3) = {*}   L(z4) = {/}
    L(z1) ∩ L(z2) ∩ L(z3) ∩ L(z4) = ∅
    
4.  <F> ::= f | g 
    f1 = f     f2 = g
    L(f1)= {f}   L(f2) = {g}
    L(f1) ∩ L(f2) = ∅
    
5.  <P> ::= a | b | c | d 
    p1 = a        p2 = b        p3 = c        p4 = d
    L(p1) = {a}   L(p2) = {b}   L(p3) = {c}   L(p4) = {d}
    L(p1) ∩ L(p2) ∩ L(p3) ∩ L(p4) = ∅
```

#### 5. Модифицированная грамматика

Так как грамматика удовлетворяет условию однозначности ветвления по первому символу, то модификация не требуется.
