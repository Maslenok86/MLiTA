#### 2. КС-грамматика языка

```
<V> ::= <F>(<V>)<D> | <P><D>
<D> ::= <Z><V> | Λ 
<Z> ::= + | - | * | / 
<F> ::= f | g 
<P> ::= a | b | c | d 
```

#### 4. Проверка однозначности ветвления по первому символу
```
1.  <V> ::= <F>(<V>)<D> | <P><D>
    v1 = <F>(<V>)<D>    v2=<P><D>
    L(v1) = {f, g}      L(v2) = {a, b, c, d}
    L(v1) ∩ L(v2) = ∅
    
2.  <D> ::= <Z><V> | Λ 
    d1 = <Z><V>             d2 = Λ
    L(d1) = {+, -, *, /}    L(d2) = {Λ}
    L(d1) ∩ L(d2) = ∅
    
3.  <Z> ::= + | - | * | / 
    z1 = +        z2 = -        z3 = *        z4 = /
    L(z1) = {+}   L(z2) = {-}   L(z3) = {*}   L(z4) = {/}
    L(z1) ∩ L(z2) ∩ L(z3) ∩ L(z4) = ∅
    
4.  <F> ::= f | g 
    f1 = f     f2 = g
    L(f1)= {f}   L(f2) = {g}
    L(f1) ∩ L(f2) = ∅
    
5.  <P> ::= a | b | c | d 
    p1 = a        p2 = b        p3 = c        p4 = d
    L(p1) = {a}   L(p2) = {b}   L(p3) = {c}   L(p4) = {d}
    L(p1) ∩ L(p2) ∩ L(p3) ∩ L(p4) = ∅
```

#### 9. Исходный код программы
```C++
#include <iostream>
#include <cstring>

using namespace std;

bool F(char ch);

bool P(char ch);

bool Z(char ch);

bool V();

bool D();

int main() {
    string ch;
    cout << "Enter your expression: ";
    if (V()) {
        cout << "Ok";
    } else
        cout << "Not Ok";
    return 0;
}

bool F(char ch) {
    char f[] = "fg";
    if (strchr(f, ch) != nullptr) {
        return true;
    } else
        return false;
}

bool P(char ch) {
    char p[] = "abcd";
    if (strchr(p, ch) != nullptr) {
        return true;
    } else
        return false;
}

bool Z(char ch) {
    char z[] = "+-*/";
    if (strchr(z, ch) != nullptr) {
        return true;
    } else
        return false;
}

bool V() {
    char ch;
    ch = cin.get();
    if (F(ch)) {
        ch = cin.get();
        if (ch == '(') {
            if (!V()) {
                return false;
            }
            if (D()) {
                return true;
            } else
                return false;
        } else
            return false;
    } else if (P(ch)) {
        if (D()) {
            return true;
        } else
            return false;
    } else
        return false;
}

bool D() {
    char ch;
    ch = cin.get();
    if (Z(ch)) {
        if (!V()) {
            return false;
        } else
            return true;
    } else {
        if ((ch == '\n') || (ch == ')') || (ch == '\0')) {
            return true;
        } else
            return false;
    }
}
```
